```python
nbits = 3  # number of hyperplanes and binary vals to produce
d = 70 # vector dimensions
```

```python
import numpy as np
# create a set of 4 hyperplanes, with 2 dimensions
plane_norms = np.random.rand(nbits, d) - .5
plane_norms
```

```python
a=np.random.random((1, 70))
b=np.random.random((1, 70))
c=np.random.random((1, 70))
# Repeat the array to generate a 1x70 array
print(a)
print(b)
dataset=[a,b,c]
print(dataset)
```

```python
# calculate the dot product for each of these
a_dot = np.dot(a, plane_norms.T)
b_dot = np.dot(b, plane_norms.T)
c_dot = np.dot(c, plane_norms.T)
a_dot
```

```python
# we know that a positive dot product == +ve side of hyperplane
# and negative dot product == -ve side of hyperplane
a_dot = a_dot > 0
b_dot = b_dot > 0
c_dot = c_dot > 0
a_dot
```

```python
# convert our boolean arrays to int arrays to make bucketing
# easier (although is okay to use boolean for Hamming distance)
a_dot = a_dot.astype(int)[0]
b_dot = b_dot.astype(int)[0]
c_dot = c_dot.astype(int)[0]
a_dot
```

```python
vectors = [a_dot, b_dot, c_dot]
buckets = {}
i = 0

for i in range(len(vectors)):
    # convert from array to string
    hash_str = ''.join(vectors[i].astype(str))
    print(hash_str)
    # create bucket if it doesn't exist
    if hash_str not in buckets.keys():
        buckets[hash_str] = []
    # add vector position to bucket
    buckets[hash_str].append(i)

print(buckets)
```

```python
 #! for tesing input query

query=np.random.random((1, 70))
print(query)

query_dot = np.dot(query, plane_norms.T)
query_dot = query_dot > 0
query_dot = query_dot.astype(int)[0]

# Convert the query array to a string
query_hash_str = ''.join(query_dot.astype(str))

# Check which bucket the query belongs to
if query_hash_str in buckets.keys():
    bucket_containing_query = buckets[query_hash_str]
    print("Query belongs to bucket:", bucket_containing_query)
else:
    print("Query doesn't match any existing buckets")
    print(0)

```
