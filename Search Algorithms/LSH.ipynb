```python
import numpy as np
import sys
import matplotlib.pyplot as plt
```

```python
dim=2
n_plans=2
# Random Planes
plane_norms=np.random.rand(n_plans,dim)-0.5

# Create a figure and axis for the plot
fig, ax = plt.subplots()

# Set the axis limits
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)


# Plot the lines passing through the origin
for i, normal in enumerate(plane_norms):
    # Calculate the direction vector of the line
    direction = np.array([-normal[1], normal[0]])
    
    # Generate points along the line
    t = np.linspace(-5, 5, 100)  # Parameter values
    x = t * direction[0]
    y = t * direction[1]
    

     # Plot the line
    # line_color = line_colors[i]
    label_x = x[50]  # Adjust the label position along the line
    label_y = y[50]
    line_label = f'v {i+1}'
    # ax.plot(x, y, label=line_label)
    ax.text(label_x, label_y, f'v {i+1}')


    
    # Add normal vector as an arrow
    scale_factor = 1  # Adjust the length of the arrow for visualization
    ax.arrow(0, 0, normal[0] * scale_factor, normal[1] * scale_factor, head_width=0.2, head_length=0.2)


# Add labels and legend
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()

# Show the plot
plt.grid()
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

```python
a=[1,2]
b=[2,1]
c=[3,1]

# Check location of a from teh 4 hyper plans
a_dot=(np.dot(a,plane_norms.T)>0).astype(int)
print(a_dot) 
b_dot=(np.dot(b,plane_norms.T)>0).astype(int)
print(b_dot) 
c_dot=(np.dot(c,plane_norms.T)>0).astype(int)
print(c_dot) 
# [-0.06920586 -0.06181368 -0.58965119  0.68051393] 0001


# Create a figure and axis for the plot
fig, ax = plt.subplots()

# Set the axis limits
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)


# Plot the lines passing through the origin
for normal in plane_norms:
    # Calculate the direction vector of the line
    direction = np.array([-normal[1], normal[0]])
    
    # Generate points along the line
    t = np.linspace(-100, 100, 100)  # Parameter values
    x = t * direction[0]
    y = t * direction[1]
    

     # Plot the line
    # line_color = line_colors[i]
    ax.plot(x, y, label=f'Normal Vector: {normal}')
    
    # Add normal vector as an arrow
    scale_factor = 1  # Adjust the length of the arrow for visualization
    ax.arrow(0, 0, normal[0] * scale_factor, normal[1] * scale_factor, head_width=0.2, head_length=0.2)

x_points = [a[0],b[0],c[0]]
y_points = [a[1],b[1],c[1]]
# Create a scatter plot for the points on the existing ax
ax.scatter(x_points, y_points, color='red', marker='o', label='Points')


labels=['a','b','c']
# Add labels next to each point
for i, label in enumerate(labels):
    ax.annotate(label, (x_points[i], y_points[i]), textcoords="offset points", xytext=(10, 10), ha='center')


# Add a legend for the points
ax.legend()

# Show the plot
ax.grid()
plt.show()
```
